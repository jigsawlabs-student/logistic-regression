## Introduction

In the last lesson, we saw how a logistic regression makes predictions through it's decision boundary.  When we represent the decision boundary as a line, if an observation falls on one side of the decision boundary, we predict it to fall into one class and on the other side we predict our observation to fall into the other class.

```python
import pandas as pd
import json
import plotly.graph_objects as go

with open('./fig.json') as json_file:
    data = json.load(json_file)
    
obj = json.loads(data)
fig = go.Figure(obj)
fig
```

> Above, we predict each observation to the left of the line to be cancerous.

We also represented our decision boundary as function, with positive outputs representing a 1, and negative outputs representing a 0.

```python
def decision_boundary_function(mean_area, mean_concavity):
    return -0.00802333*mean_area - 3.47361781 * mean_concavity + 5.86599622
```

## Adding Probability

Now we'd like to have our logistic regression function do more than just making a prediction of a 1 if the output is positive, or 0 if negative.  We  also care about *how* positive or negative the decision function's return value is.  After all, notice that the further an observation is from the decision boundary, the less likely we are to get it wrong.

```python
fig
```

Hovering over datapoints in the above graph, also notice that, the further away an observation is from the decision boundary, the further the output is from zero.

So now, what we would like to do is turn these values into our scale of 1 or 0.  And we want the high negative values to approach 0 and the high positive values to approach 1.   

### Introducing the sigmoid function

As it turns out, our sigmoid function does a great job of mapping our positive and negative values to probabilities between 0 and 1.  Let's learn about it.

The sigmoid function, $\sigma$, is defined as the following: 

$$ \sigma(x) = \frac{1}{1 + e^{-x}} $$ 

Here, $x$ is the score that is returned from our decision function.

Now, let's review our goals, we wanted a function that did the following: 

* transformed our large negative scores to a value close to zero 
* transformed our large positive scores to a value close to one
* equaled one half when $x = 0$ 

Let's see how it does:

```python
import numpy as np
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
sigmoid(3)
```

```python
sigmoid(-3)
```

```python
sigmoid(0)
```

Ok, so now let's see how our sigmoid function performs.

1. When $x = 0$ we have:

$$ \sigma(0) = \frac{1}{1 + e^{-0}} = \frac{e^0}{e^0 + 1} = \frac{1}{1 + 1} = \frac{1}{2}$$ 

2. When $x = -100$ (a somewhat large negative number) we have:

$$ \sigma(-100) = \frac{1}{1 + e^{- (-100)}} = \frac{1}{1 + e^{100}} \approx \frac{1}{1 + \infty} \approx \frac{1}{\infty} \approx 0 $$ 

2. When $x = 100$ (a somewhat large positive number) we have:

$$ \sigma(-100) = \frac{1}{1 + e^{-100}} = \frac{1}{1 + \frac{1}{e^{100}}} = \frac{1}{1 + 0} \approx 1 $$ 

So we can see that this function appears to meet our conditions, and would work well for us.  Now when we have a decision function, it can take the distances of the points from the decision function.  If the distance is far from the line in the positive direction, we return a probability close to one.  If it's far from the line in the negative direction, we return a probability close to zero.  And if the distance is near the line we return a probability close to fifty percent. 

### From decision boundary plot to sigmoid plot

Ok, let's go back to our original problem of identifying cancer cells.  Remember that our goal is to find a decision function that differentiates between benign and cancerous cell.  This is our plot.

```python
fig
```

So now let's transform the outputs from our decision function into probabilities.

> To do so, we first download the data.

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
dataset = load_breast_cancer()
cols = ['mean area', 'mean concavity']

X = pd.DataFrame(dataset['data'], columns = dataset['feature_names'])[cols]
y = pd.Series(dataset['target'])
```

And train our model.

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver = 'liblinear')
model.fit(X, y)
```

So that we can now get outputs from the decision function.

```python
model.decision_function(X)[:3]
```

And we can transorm these into probabilities with the sigmoid function.

```python
import numpy as np
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
vals = model.decision_function(X[:3])
[sigmoid(val) for val in vals ]
```

Or we can use the logistic regression model's predic_proba function to get the same result.

```python
model.predict_proba(X[:3])
```

> The probability of a zero is in the leftmost column, and the probability of a 1 is in the rightmost column.

### Summary

In this lesson, we saw how a logistic regression model takes the output from it's decision function, with values between positive infinity and negative infinity, and translates it into values between 0 and 1.  The way it does so is through the sigmoid function, represented by the greek letter, sigma:

$$ \sigma(x) = \frac{1}{1 + e^{-x}} $$ 

The sigmoid function maps values approaching negative infinity close to 0, and values approaching positive infinity close to 1.  If the output from our decision function is a 0, the sigmoid function translates it to .5.
