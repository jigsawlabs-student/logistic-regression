# Logistic Regression Hypothesis Function

### Introduction

With logistic regression we are predicting whether an event will occur or not.  Many problems are framed this way -- some either joins a company or does not, clicks an email or does not, clicks subscribe or does not.  With logistic regression we can predict the outcome of an event, predict the probability that an outcome will occur, and uncover the features are predictive of a particular outcome.  Let's get started.


### Introducing our Problem

Let's introduce ourselves to logistic regression by working with the famous breast cancer dataset.  The dataset was formed by researchers looking at different slides of patient cells, and for each slide the researchers indicated features of the cells on the slide (like the mean area), as well as if cancerous cells were present on the slide, the target.  

<img src="./cancer-dataset.png" width="20%">

Let's load up our dataset and get started.

```python
from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer()
```

From there, we'll assign our features to the variable `X` and the target to `y`.

```python
import pandas as pd
X = pd.DataFrame(dataset['data'], columns = dataset['feature_names'])
y = pd.Series(dataset['target'])
```

```python
X[:3]
```

If we look at the feature names, we can see some of the features that were observed.  

```python
X.shape
```

### Visualizing our problem

Let's begin by plotting two features of our dataset, along with the associated outcomes.

We'll begin by selecting those observations where cancer was identified, and those where it was not.

```python
cancerous_df = X[y == 1]
benign_df = X[y == 0]
```

Now let's take a look at the concavity and mean area of the related cells.

```python
mean_area_cancerous = cancerous_df['mean area']
mean_area_benign = benign_df['mean area']
concavity_cancerous = cancerous_df['mean concavity']
concavity_benign = benign_df['mean concavity']
```

Next, let's plot our data.

```python
import plotly.graph_objects as go

cancerous_scatter = go.Scatter(x = mean_area_cancerous, y = concavity_cancerous, 
                               name = 'cancerous', mode = 'markers')

benign_scatter = go.Scatter(x = mean_area_benign, y = concavity_benign,  name = 'benign', mode = 'markers')
layout = {'xaxis': {'title': 'mean area'},
          'yaxis': {'title': 'mean concavity', 
                    'range': [0, .5]}, 'title': 'Is it cancerous?'}
go.Figure([cancerous_scatter, benign_scatter], layout = layout)
```

We can see from the above plot that there is a trend in the data.  In fact, if we were going to infer a trend in the data, it would may be to draw a line dividing the data.  If the data falls on one side of the line, we predict the observation to be cancerous, and otherwise we predict the observation is benign.  This dividing line is called the logistic regression model's **decision boundary**.   Let's train a model and see it action.

### Training a Model

We'll train a logistic regression model by fitting the model along with the `mean area` and `mean concavity` features, and the target variable.

```python
from sklearn.linear_model import LogisticRegression
scoped_df = X[['mean area', 'mean concavity']]

model = LogisticRegression()
model.fit(scoped_df, y)
```

Then, after fitting our logistic regression model, we are given a set of parameters.  Let's take a look.

```python
model.coef_
```

```python
model.intercept_
```

We can plot our decision boundary by calling the `decision_boundary_y` function below above with a range of values.

> It's ok if you do not understand the code below.  What's more important is the figure that follows.

```python
area_values = list(range(500, 1000, 5))

def decision_boundary_y(x, coefs, intercept):
    coef_pair = coefs[0]
    return (coef_pair[0]/coef_pair[1])*x + intercept[0]/coef_pair[1]

boundaries = [decision_boundary_y(idx, model.coef_, model.intercept_) for idx in area_values]
```

```python
import plotly.graph_objects as go
def decision_boundary_function(mean_area, mean_concavity):
    return -0.00802333*mean_area - 3.47361781 * mean_concavity + 5.86599622

cancerous_predictions = [decision_boundary_function(area, concavity)for area, concavity in zip(mean_area_cancerous, concavity_cancerous)]
benign_predictions = [decision_boundary_function(area, concavity) for area, concavity in zip(mean_area_benign, concavity_benign)]

cancerous_scatter = go.Scatter(x = mean_area_cancerous, y = concavity_cancerous, 
                               name = 'cancerous', mode = 'markers', hovertext = cancerous_predictions)

benign_scatter = go.Scatter(x = mean_area_benign, y = concavity_benign, 
                            hovertext = benign_predictions, name = 'benign', mode = 'markers')
decision_trace = go.Scatter(x=area_values, y=boundaries,
                            mode = 'lines', name = '-3.47*x_1 -0.008*x_2  = 0')

layout = {'xaxis': {'title': 'mean area'}, 'yaxis': {'title': 'mean concavity', 'range': [0, .5]}, 'title': 'Is it cancerous?'}
fig = go.Figure([cancerous_scatter, benign_scatter, decision_trace], layout = layout)
```

```python
fig
```

Let's take a look at the diagram above.  The green line is the fitted decision boundary.  This was derived from the parameters that the model provided us above.

```python
model.coef_, model.intercept_
```

We'll discuss *how* the model came to the decision boundary when we discuss the cost function.  But for now, let's just assume that this boundary is correct.

The decision boundary indicates the predictions of our model.  Everything to the right of the line is predicted as benign, and everything to the left of the line is predicted as cancerous.

Now that we understand how the hypothesis function works, we can see that there are a half dozen orange cancerous points on the benign side of the line.  And there are many more benign observed cells on the cancerous side of our decision boundary.  That's ok, we don't expect our model to be perfect in it's predictions.

### Our predictions mathematically

So we saw how we can use our decision boundary to visually see if our model predicts a value as being cancerous or non-cancerous.  We can also use our trained parameters to predict our outcomes.

Let's take another look at our model parameters.

```python
model.coef_
```

```python
model.intercept_
```

We can use these parameters to define the following function. 

```python
def decision_boundary_function(mean_area, mean_concavity):
    return -0.00802333*mean_area - 3.47361781 * mean_concavity + 5.86599622
```

Then we can pass through data from different observations.  If our function outputs a positive number, we'll predict our datapoint as being one, or cancerous.  If our function outputs a negative number, then we'll interpret our datapoint as 0, or benign.

```python
X[['mean area', 'mean concavity']][17:20]
```

```python
decision_boundary_function(798.8, 0.17220)
```

Lucky for us, sklearn gives us this decision function for free.  

```python
model.decision_function(X[['mean area', 'mean concavity']])[17:20]
```

And we can even turn these outputs into predictions of 1 or 0.  Notice that everything less negative receives a prediction of 0, and positive number get a prediction of 1.

```python
model.predict(scoped_df[['mean area', 'mean concavity']])[17:23]
```

And note that on this sample, our model did well predicting this snippit of data.

```python
updated_df = scoped_df.assign(y = y)
updated_df.loc[17:22, :]
```

### Summary

In this lesson, we learned about the classification problem and how we can use logistic regression to predict whether an observation is likely to fall into one group or another.  We saw that with a classification problem, the values of our observed labels are discrete.  In our example, they were either one or zero to indicate a cancerous cell or not.  We use the features to fit a decision function that predicts if an observation is positive or negative.  If we pass the feature variables of an observation into a decision function, the sign of the return value is used to make a positive (cancerous) or negative observation.  That is, when our decision function returns a positive value, we predict a value of one, and when our decision function returns a negative value of zero.
