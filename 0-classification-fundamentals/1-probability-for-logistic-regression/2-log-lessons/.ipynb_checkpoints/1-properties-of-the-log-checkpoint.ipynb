# Properties of the Log

### Introduction

In the last lesson, we learned some of the basic mechanics of the logarithmic function.  We saw that we can define the log generally as: 

> if $b^y = x$, then $\log_b(x) = y$.

So if $2^x = 32$ then $\log_2(32) = 5$

We also saw that we oftentimes use the natural log, $\ln$, which uses a base of $e$ instead of 2.  In this lesson, we'll learn some additional properties of the logarithmic function.

### Properties of the Log

Let's notice something about the $log_2(x)$.

```python
import numpy as np
np.log2(64)
```

```python
np.log2(4096)
```

So we just increased our value $x$ by over 4000, but the log of only increased by 6.  We can see this visually if we plot the $\log(x) $ for a range of numbers.

```python
import numpy as np
x_vals = np.arange(.1, 100, .1)
```

```python
log_x = np.log(x_vals)
```

```python
import plotly.graph_objects as go

scatter = go.Scatter(x = x_vals, y = log_x)
go.Figure(scatter, {'title': '$ln(x)$'})
```

So we can see from the plot above, that the log dramatically decreases larger numbers.  This is what leads to the plateau we see in the plot above.

Another key point is that as $x$ increases, $y$ also increases.  Always.  In other words, 

* if $ a > b $, then $ln(a) > ln(b) $.

This is called a monotonic function.  Here is a function that is not monotonic.

```python
x_vals = np.arange(-10, 10, .1)
scatter = go.Scatter(x = x_vals, y = x_vals**2)
go.Figure(scatter, {'title': '$x^2$'})
```

Do you see why it's not monotonic?  We can see that:
* $-5 \gt-6$, but 
* $25 \lt 36$.

> In the future, we'll make use of the monotonic nature of the logarithmic function.

### Simplifying functions with the log

As we'll see, many functions make use of the log.  The reason, believe it or not, is that applying the log to a function can *simplify* the math.  Here are some properties of the log that help equations easier:

$\log(ab) = \log(a) + \log(b)$

```python
np.log2((2**6) * (2**7))
```

```python
np.log2(2**6)  +  np.log2(2**7)
```

> So applying the log to a product, allows us to use addition instead of multiplication.

$log(a^b) = b* log (a)$

```python
np.log2(2**8)
```

```python
8*np.log2(2)
```

> Applying the log to an exponent reduces our problem to a product.

And the derivative of the log isn't so bad either.

$f(x) = log(x)$, then 

$f'(x) = \frac{1}{x}$

### Summary

In this lesson we reviewed some properties of the log.  We saw the following:

* Applying the log to a number, the log reduces large numbers much much more than small numbers.
* The log is a monotonic function.  So if $ a > b $, then $ln(a) > ln(b) $.

We also saw that the log can be used to simplify equations:

* It reduces products to sums:
    * $\log(ab) = \log(a) + \log(b)$
    
* It reduces exponents to products:
    * $log(a^b) = b* log (a)$
    
* If $f(x) = log(x)$ then:
    * $f'(x) = \frac{1}{x}$.
